### Task 2: Build Advanced Preprocessing Function
- Implement the following function:
- def preprocess_text(text):
-     pass

```python
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    text = text.lstrip()
    text = text.rstrip()
    text = text.translate(str.maketrans("","",string.punctuation))
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    tokens = text.split()
    filtered_tokens = [word for word in tokens if len(word) >= 2 or word in ['i', 'no', 'not']]
    return filtered_tokens
```

```python
text = ["I have 2 dogs",
"This is good",
"soooo goooood!!!", 
 "WOW!!! This is GREAT!!!",
"Visit http://example.com now"]
for i in text:
    print(f'Processed: {preprocess_text(i)}')
```

```python
text = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
for i in text:
    print(f"\n Original Text: {i}")
    res=preprocess_text(i)
    print(f"Cleaned Tokens: {res}")
    print("Sentence: ", " ".join(res))
```

### Task 4: Token Analytics
- For each processed sentence, compute:
- Total number of tokens
- Number of unique tokens
- Average token length

```python
all_tokens = []
for i in text:
    print(f"\n Original Text: {i}")
    res = preprocess_text(i)
    all_tokens.extend(res)
    print(f"Cleaned Tokens: {res}")
    print("Total number of tokens:", len(res))
    print("Number of unique tokens: ", len(set(res)))
    avg_len = sum(len(t) for t in res) / len(res) if len(res) > 0 else 0
    print("Average token length: ", avg_len)
    print("Sentence: ", " ".join(res))
```

### Task 6: Build Full Pipeline
- def full_pipeline(text_list):
-         pass 

#### Expected Output
- {
-     "tokens": [...],
-     "clean_sentences": [...]
- }

```python
def full_pipeline(text_list):
    clean_sentences = []
    all_tokens = []
    for i in text_list:
        res = preprocess_text(i)
        all_tokens.extend(res)
        clean_sentences.append(" ".join(res))
    return {
        "tokens": all_tokens, 
        "clean_sentences": clean_sentences
    }
```
